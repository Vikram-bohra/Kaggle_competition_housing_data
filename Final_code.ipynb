```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X = data_train.values
y = price.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python

# Applying PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 50)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(n_estimators= 300)

reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,explained_variance_score
acc = mean_squared_error(y_test,pred)
r_value = r2_score(y_test,pred)
var_score = explained_variance_score(y_test,pred)
```

```python
import sklearn.metrics as sm
#pcs features = 50
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test, pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, pred), 2)) 
print("R2 score =", round(sm.r2_score(y_test, pred), 2))
print("root mean squared log error",round(sm.mean_squared_log_error(y_test,pred), 2))
```
